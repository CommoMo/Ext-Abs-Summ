```python
import json, os
from tqdm import tqdm
```

```python
data_types = ['train', 'valid']
dataset_name = 'article'
save_path = '/workspace/data/'
```

```python
def get_processed_dataset(raw_dataset):
    docs = raw_dataset['documents']
    result_list = []
    for doc in tqdm(docs):
        temp_dict = {}
        ext_indices = doc['extractive']

        abs = doc['abstractive'][0]
        exts = [sen['sentence'] for para in doc['text'] for sen in para if sen['index'] in ext_indices]
        sentences = [sen['sentence'] for para in doc['text'] for sen in para]
        
        if len(exts) ==0 or len(sentences) <= 2 or len(abs) <= 5:
            print('passed')
            print(exts)
            print(sentences)
            print(abs)
            continue

        temp_dict['ext'] = exts
        temp_dict['abs'] = abs
        temp_dict['sentences'] = sentences
        result_list.append(temp_dict)
    return result_list

for data_type in data_types:
    data_path = f'../data/{dataset_name}/{data_type}_original.json'
    raw_dataset = json.load(open(data_path, 'r', encoding='utf-8'))
    processed_dataset = get_processed_dataset(raw_dataset)
    with open(os.path.join(save_path, dataset_name, f'{data_type}_dataset.json'), 'w', encoding='utf-8') as f:
        json.dump(processed_dataset, f, ensure_ascii=False, indent='\t')
    
```
